```python
orbits = dict()
for line in open('inputs/day6.txt'):
    com, planet = line.strip().split(")")
    assert planet not in orbits
    orbits[planet] = com
    
    
def get_number_orbits(planet, orbit_counter, planets):
    if planet not in orbit_counter:
        orbit_counter[planet] = get_number_orbits(orbits[planet], orbit_counter, planets) + 1
    return orbit_counter[planet]

orbit_counter = {"COM": 0}
total_orbits = 0
for planet in orbits.keys():
    total_orbits += get_number_orbits(planet, orbit_counter, orbits)
print('solution part 1', total_orbits)

def get_orbit_chain(planet, orbits):
    if planet == 'COM':
        return []
    return [orbits[planet]] + get_orbit_chain(orbits[planet], orbits)
orbit_you = get_orbit_chain('YOU', orbits)
orbit_san = get_orbit_chain('SAN', orbits)

steps_to_common_you = 0
steps_to_common_san = 0
for planet in orbit_you:
    if planet in orbit_san:
        common_orbit = planet
        break
    steps_to_common_you += 1
        
for planet in orbit_san:
    if planet == common_orbit:
        break
    steps_to_common_san += 1

print('solution part 2:', steps_to_common_you + steps_to_common_san)
```
